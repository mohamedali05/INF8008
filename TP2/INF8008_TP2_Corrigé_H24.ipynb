
### A)
À l'aide de Pandas, anonymisez les données chargées au niveau de la colonne 'Id' dans une variable nommée 'df'. Triez le jeu de données par Id et Time, puis affichez le nouveau DataFrame. <b>(2 points)</b>

```python
#TODO:
df['Id'] = pd.factorize(df['Id'])[0].astype(float)
df = df.sort_values(by=['Id', 'Time'])
df
```


### B)
Vous remarquerez que dans ce jeu de données, les positionnements consécutifs ne sont pas particulièrement informatifs.

Réduisez la taille des données en ne gardant que le dernier positionnement. Un positionnement consécutif a lieu si pour deux lignes consécutives, l'utilisateur et le module restent les mêmes et la valeur de Par2 indique 'position' au début.

Vous devez donc :
1. Créer une colonne "pos" qui indique si la valeur de "Par2" commence par 'position';
2. Créer une colonne "pos.fin" qui indique s'il s'agit d'une position de fin au niveau de la colonne "pos"
3. Créer un dataframe qui n'inclue pas les positionnements répétitifs

Affichez le nouveau dataframe. Vous devriez avoir autour de 18068 lignes dans celui-ci. Vous devez utiliser la fonction ci-dessous pour vous aider. <b>(4 points)</b>

```python
def seq_last(v):

    """ Renvoie un tableau de booléens indiquant si chaque élément de la séquence d'origine v est le dernier d'une séquence. """

    v_np = np.array(v)
    return (np.concatenate(([True], v_np)) & np.concatenate((~v_np, [True])))[1:]
```

```python
#TODO:

df2 = df.copy()
df2['pos'] = df2['Par2'].str.startswith('position')

df2['pos.fin'] = seq_last(df2['pos'])

df3 = df2[~df2['pos'] | df2['pos.fin']]
df3
```

**Note :** Pour la suite du travail, assurez-vous que votre jeu de données soit ordonné de manière chronologique, tel que demandé à l'étape A. En effet, les étapes C à F sont des sous-étapes pour la création des sessions.

### C)
Ajoutez une colonne nommée "nouv.u" à votre DataFrame obtenu à la question B. Celle-ci détermine si chaque ligne du DataFrame correspond à la première occurrence (non dupliquée) de la valeur dans la colonne "Id".

Affichez le nouveau DataFrame. <b>(1 points)</b>

```python
#TODO:
df3['nouv.u'] = ~df3['Id'].duplicated()
df3
```

### D)
Convertissez les valeurs de la colonne "Time" en DateTime, puis ajoutez une colonne nommée "Duree" à votre DataFrame. Celle-ci contient le calcul du temps écouler entre deux lignes du DataFrame en fonction de la colonne Time. N'oubliez pas de mettre la durée à 0 lorsqu'il s'agit d'un nouvel utilisateur (i.e., La valeur dans la colonne "nouv.u" est égale à True).

Note: Vous devrez utiliser le fonction [diff](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.diff.html) pour calculer la durée ainsi que [Timedelta](https://pandas.pydata.org/docs/reference/api/pandas.Timedelta.html) pour mettre celle-ci à 0.

Affichez le nouveau DataFrame. <b>(3 points)</b>

```python
#TODO:

df4 = df3.reset_index(drop=True)
df4['Time'] = pd.to_datetime(df4['Time'])

df5 = df4.copy()

df5['Duree'] = df4['Time'].diff().fillna(pd.Timedelta(seconds=0))

df5.loc[df5['nouv.u'], 'Duree'] = pd.Timedelta(seconds=0)

df5
```

### E)
Indiquez les évènements qui sont potentiellement des pauses. Pour ce faire, créez une colonne "pause" contenant des valeurs booléennes. Un évènement est une pause si la durée entre deux lignes est supérieure à 90 minutes et que celles-ci sont associées au même utilisateur.

Affichez le nouveau DataFrame. <b>(2 points)</b>

```python
#TODO:

# pause_lim = pd.Timedelta(90 * 60, unit='s')
pause_lim = pd.Timedelta(minutes = 90)

df6 = df5.copy()

df6['pause'] = (df5['Duree'] > pause_lim) & (~df5['nouv.u'])

df6
```

### F)
Créez une colonne nommée "Session". Celle-ci indique le numéro de session associé à chaque ligne du DataFrame.
Ce numéro est déterminé en utilisant la fonction cumsum() dans le cas où c'est un nouvel utilisateur ("nouv.u" possède une valeur True) ou bien une pause ("pause" possède une valeur True).

Affichez le nouveau DataFrame. <b>(2 points)</b>

Note: Vous devrez utiliser la fonction [cumsum](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cumsum.html) pour assigner un numéro de session.



```python
#TODO:

df7 = df6.copy()

df7['Session'] = (df7['pause'] | df7['nouv.u']).cumsum()
df7
```

### G)
Déterminez le nombre de sessions par "Id" utilisateurs.

Affichez le résultat. <b>(2 points)</b>

```python
#TODO:
session_counts = df7.groupby('Id')['Session'].nunique()
session_counts
```

### H)
Changez la valeur de Categ pour "pause" lorsqu'il s'agit d'une pause.

Combien d'entrée de chaque catégorie contient votre jeu de données (les catégories sont "ev", "ex" et "pause")? Puis, affichez le résultat. <b>(2 points)</b>

```python
#TODO:
df8 = df7.copy()

df8.loc[df8['pause'], 'Categ'] = 'pause'

categ_counts = df8['Categ'].value_counts()
# ou bien
# df8.groupby('Categ')['Categ'].count()
categ_counts
```
